# **Import Necessary Libraries**

---



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Loading the Dataset**

---



```python
df = pd.read_csv("icr_train.csv")
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

```python
df["EJ"].value_counts()
```

# **Check for NULL Values**

---



```python
df.isnull().sum()
```

```python
for i in df.columns:

  if(df[i].isnull().any() == True):

    mean = df[i].mean()
    df[i] = df[i].fillna(mean)
```

```python
df.isnull().sum().max()
```

# **Data Visualsation**

---



```python
sns.countplot(x="EJ", data=df, hue="Class")
```

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(20,20))
sns.heatmap(corr, annot=True)
```

# **Seperating Dependent & Independent Variables**

---



```python
X = df.drop(["Id", "Class"], axis=1)
y = df["Class"]
```

```python
X.head()
```

```python
y.head()
```

# **Label Encoding**

---



```python
from sklearn.preprocessing import LabelEncoder
ej_le = LabelEncoder()
```

```python
X["EJ"] = ej_le.fit_transform(X["EJ"])
```

# **Splitting into Training & Testing Set**

---



```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
                                                    random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# **Model Building - XGBClassifier**

---



```python
from xgboost import XGBClassifier
xgbc_model = XGBClassifier()
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
xgbc_params = {'gamma' :[i/10.0 for i in range(0,5)],
               'max_depth' : [2,4,6,8,10,12,13,15],
               'min_child_weight':[1,2,5,10],
               'max_delta_step':[0,1,2,5,10],
               'subsample' :[i/10.0 for i in range(5,10)],
               'colsample_bytree' :[i/10.0 for i in range(5,10)],
               'colsample_bylevel' :[i/10.0 for i in range(5,10)],
               'n_estimators' : [100,150,200,300]
               }
```

```python
xgbc_cv = RandomizedSearchCV(estimator=xgbc_model,
                            param_distributions=xgbc_params,
                            cv=10)
```

```python
xgbc_cv.fit(X_train, y_train)
```

```python
xgbc_cv.best_params_
```

```python
xgbc_model = XGBClassifier( subsample = 0.9,
                           n_estimators = 300,
                            min_child_weight = 1,
                            max_depth = 8,
                            max_delta_step = 0,
                            gamma = 0.4,
                            colsample_bytree = 0.6,
                            colsample_bylevel = 0.7
                            )
```

```python
xgbc_model.fit(X_train, y_train)
```

```python
xgbc_model.score(X_test, y_test)
```

```python
y_preds1 = xgbc_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_preds1))
```

# **Model Building - Random Forest Classifier**

---



```python
from sklearn.ensemble import RandomForestClassifier
rfc_model = RandomForestClassifier()
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rfc_params = {'max_depth': list(range(10, 15)),
               'max_features': list(range(2,14))}
```

```python
rfc_cv = RandomizedSearchCV(estimator=rfc_model,
                            param_distributions=rfc_params,
                            cv=10)
```

```python
rfc_cv.fit(X_train, y_train)
```

```python
rfc_cv.best_params_
```

```python
rfc_model = RandomForestClassifier(
    max_features=12, max_depth=11
)
```

```python
rfc_model.fit(X_train, y_train)
```

```python
rfc_model.score(X_test, y_test)
```

```python
y_preds2 = rfc_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_preds2))
```
